```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('CD.csv')
```

```python
df.info()
```

```python
df
```

```python
#Total custodial deaths per state

total_deaths_per_state = []
per_state = 0
        
for row in range(0, 38):
    per_state = 0
    for col in range(1, len(df.columns), 4):
        #print(df.iloc[row, col])
        per_state += df.iloc[row, col]
    total_deaths_per_state.append(per_state)
        
print(total_deaths_per_state)        
```

```python
len(total_deaths_per_state)
```

```python
x = pd.Series(total_deaths_per_state)
df['TOTAL CUSTODIAL DEATHS'] = x.values
df.head()
```

```python
#Total cases registered per state

total_cases_reg_per_state = []
per_state = 0
        
for row in range(0, 38):
    per_state = 0
    for col in range(2, len(df.columns), 4):
        #print(df.iloc[row, col])
        per_state += df.iloc[row, col]
    total_cases_reg_per_state.append(per_state)
        
print(total_cases_reg_per_state) 
```

```python
x = pd.Series(total_cases_reg_per_state)
df['TOTAL CASES REGISTERED'] = x.values
df.head()
```

```python
#Total policemen chargesheeted per state

total_police_chargesheet_per_state = []
per_state = 0
        
for row in range(0, 38):
    per_state = 0
    for col in range(3, len(df.columns), 4):
        #print(df.iloc[row, col])
        per_state += df.iloc[row, col]
    total_police_chargesheet_per_state.append(per_state)
        
print(total_police_chargesheet_per_state) 
```

```python
x = pd.Series(total_police_chargesheet_per_state)
df['TOTAL POLICEMEN CHARGESHEETED'] = x.values
df.head()
```

```python
#Total policemen convicted per state

total_police_convicted_per_state = []
per_state = 0
        
for row in range(0, 38):
    per_state = 0
    for col in range(4, len(df.columns), 4):
        #print(df.iloc[row, col])
        per_state += df.iloc[row, col]
    total_police_convicted_per_state.append(per_state)
        
print(total_police_convicted_per_state) 
```

```python
x = pd.Series(total_police_convicted_per_state)
df['TOTAL POLICEMEN CONVICTED'] = x.values
df.head()
```

```python
df.to_csv("CD_Updated.csv", index = False)
```

## Notes for visualization

1. A bar graph showing how many deaths have occurred, cases have been registered, police have been convicted etc. in total
2. A bar graph showing distribution of deaths in total per state

## 1. A bar graph showing how many deaths have occurred, cases have been registered, police have been convicted etc. in total

```python
x1_values = list(["CUSTODIAL DEATHS", "CASES REGISTERED", "COPS CHARGESHEETED", "COPS CONVICTED"])
y1_values = list(df.iloc[37, -4:])

# Figure Size 
fig, ax = plt.subplots(figsize =(16, 9)) 
  
# Horizontal Bar Plot 
ax.barh(x1_values, y1_values, color = ["#cccc00", "#6666ff", "#ff9933", "#ff6699"]) 
  
# Remove axes splines 
for s in ['top', 'bottom', 'left', 'right']: 
    ax.spines[s].set_visible(False) 
  
# Remove x, y Ticks 
ax.xaxis.set_ticks_position('none') 
ax.yaxis.set_ticks_position('none') 
  
# Add padding between axes and labels 
ax.xaxis.set_tick_params(pad = 5) 
ax.yaxis.set_tick_params(pad = 10) 
  
# Add x, y gridlines 
ax.grid(b = True, color ='grey', 
        linestyle ='-.', linewidth = 0.5, 
        alpha = 0.2) 
  
# Show top values  
ax.invert_yaxis() 
  
# Add annotation to bars 
for i in ax.patches: 
    plt.text(i.get_width()+0.2, i.get_y()+0.5,  
             str(round((i.get_width()), 2)), 
             fontsize = 10, fontweight ='bold', 
             color ='grey') 


# Add title 
ax.set_title('Statistics Regarding Custodial Deaths in India 2001-12', 
             loc ='left', ) 
  
# Add Text watermark 
fig.text(0.9, 0.15, 'By Sthitaprajna Mishra', fontsize = 12, 
         color ='grey', ha ='right', va ='bottom', 
         alpha = 0.7) 
  
# Show Plot 
plt.show()  
```

## 4. A line graph showing total death distribution over all the years vs total cops convicted over the years(the strong disparity shows here that hardly any cops are punished)

```python
x4_values = [i for i in range(2001, 2013)]
x4_values
```

```python
final_df.head()
```

```python
y4_1_values = []

for i in range(1, 48, 4):
    y4_1_values.append(int(final_df.iloc[37, i]))
        
print(y4_1_values) 
```

```python
y4_2_values = []

for i in range(4, 52, 4):
    y4_2_values.append(int(final_df.iloc[37, i]))
        
print(y4_2_values) 
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
```

```python
# plotting the line 1 points 
plt.plot(x4_values, y4_1_values, label = "Deaths", color="#ff9900", linestyle='dashed', linewidth = 2, 
         marker='o', markerfacecolor='#ff3300', markersize=8)

# plotting the line 2 points 
plt.plot(x4_values, y4_2_values, label = "Convicted", color= "#6666ff", linestyle='dashed', linewidth = 2, 
         marker='o', markerfacecolor='blue', markersize=8)

# Add x, y gridlines 
plt.grid(b = True, color ='grey', 
        linestyle ='-.', linewidth = 0.5, 
        alpha = 0.2)  

plt.title('Custodial Deaths vs Cops Convicted 2001-12')
plt.legend()
```
